## Objective
To analyze and visualize sentiment patterns in textual data to understand
public opinion using basic sentiment analysis techniques.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from textblob import TextBlob
```

```python
data = {
    "text": [
        "I love this product, it works amazing!",
        "This is the worst service ever.",
        "Very happy with the experience.",
        "I am disappointed and unhappy.",
        "Absolutely fantastic quality!",
        "Not worth the money.",
        "Great customer support.",
        "Terrible experience overall."
    ]
}

df = pd.DataFrame(data)
df
```

```python
def get_sentiment(text):
    analysis = TextBlob(text)
    if analysis.sentiment.polarity > 0:
        return "Positive"
    elif analysis.sentiment.polarity < 0:
        return "Negative"
    else:
        return "Neutral"

df['Sentiment'] = df['text'].apply(get_sentiment)
df
```

```python
sns.countplot(x='Sentiment', data=df)
plt.title("Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.show()
```

## Conclusion
The sentiment analysis shows a mix of positive and negative opinions.
Positive sentiments slightly dominate, indicating overall favorable feedback.
This task demonstrates how basic NLP techniques can be used to understand
public opinion from text data.
